```python
import MDAnalysis as md
import numpy as np
import pandas as pd
from MDAnalysis.analysis.rms import rmsd
```

```python
gen_enc_k = md.Universe('../ionic_strength/frame1_gen_enc_k_165mM.gro', '../ionic_strength/generated_encodermap_k_165mM.xtc')   
traj_k = md.Universe('../ionic_strength/frame1_gen_enc_k_165mM.gro', '../ionic_strength/k_5_165mM_md_500ns_protein.trr')

gen_enc_wt = md.Universe('../ionic_strength/frame1_gen_enc_wt_150mM.gro', '../ionic_strength/generated_encodermap_wt_150mM.xtc')   
traj_wt = md.Universe('../ionic_strength/frame1_gen_enc_wt_150mM.gro', '../ionic_strength/wt_8_150mM_md_500ns_protein.trr')
```

```python
frames_comparison_k = []

for min_str in gen_enc_k.trajectory:
    frames_comparison_k.append([rmsd(traj_frame.positions, min_str.positions) for traj_frame in traj_k.trajectory])  

frames_min_original_trajectory_k = np.array(frames_comparison_k).argmin(axis=1) + 1
```

```python
mapping_k = pd.DataFrame({'generated encodermap': range(1, 33), 
                          'frames trajectory trr': frames_min_original_trajectory_k})

# Assign closest frame instead of same trajectory frame in order to have 32 different structures
mapping_k.iloc[2]['frames trajectory trr'] = 16
mapping_k.iloc[5]['frames trajectory trr'] = 242
mapping_k.iloc[6]['frames trajectory trr'] = 243
mapping_k.iloc[16]['frames trajectory trr'] = 26
mapping_k.iloc[19]['frames trajectory trr'] = 158

mapping_k
```

```python
mapping_k.sort_values('frames trajectory trr', ignore_index=True)
```

```python
frames_comparison_wt = []

for min_str in gen_enc_wt.trajectory:
    frames_comparison_wt.append([rmsd(traj_frame.positions, min_str.positions) for traj_frame in traj_wt.trajectory])  

frames_min_original_trajectory_wt = np.array(frames_comparison_wt).argmin(axis=1) + 1
```

```python
mapping_wt = pd.DataFrame({'generated encodermap': range(1, 33), 
                          'frames trajectory trr': frames_min_original_trajectory_wt})

# Assign closest frame instead of same trajectory frame in order to have 32 different structures
mapping_wt.iloc[26]['frames trajectory trr'] = 11
mapping_wt.iloc[20]['frames trajectory trr'] = 129
mapping_wt.iloc[21]['frames trajectory trr'] = 130
mapping_wt.iloc[27]['frames trajectory trr'] = 177
mapping_wt.iloc[28]['frames trajectory trr'] = 178
mapping_wt.iloc[30]['frames trajectory trr'] = 199
mapping_wt.iloc[8]['frames trajectory trr'] = 200

mapping_wt
```

```python
mapping_wt.sort_values('frames trajectory trr', ignore_index=True)
```
